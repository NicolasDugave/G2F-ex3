## Exercise 1: Transcriptional bursting

Let's consider the transcriptional process in which an mRNA $x$ is produced at a rate $s$ and degraded at a rate $k$. We will consider a two-state model for the promoter (Random Telegraph process). This means that the promoter can switch from an active ($g=1$) to inactive state ($g=0$) and vice-versa with respective rates $k_{off}$ and $k_{on}$.


The Telegraph process has the following elementary reactions:  
\begin{array}{lll}
(1, n) \longrightarrow (1, n+1) \textrm{ with rate } s \\
(g, n) \longrightarrow (g, n-1) \textrm{ with rate } k\,n \\
(1, n) \longrightarrow (0, n) \textrm{ with rate } k_{off} \\
(0, n) \longrightarrow (1, n) \textrm{ with rate } k_{on}
\end{array}

We describe the states $(g, n)$ with $g\in \{1, 0\}$, $n \in \mathbb{N}$.

### Answers 

When the promoter is active, the reactions that the stoichiometry matrix represents are transcription rate (s), degradation rate (k*n) and the switching of the promoter to the inactive state (Koff), hence we have a column for each reaction and a row for each species (n,g) the reactions are happening to. When the promoter is inactive, transcription is not possible, so there are only two columns for the reactions of degradation and switching of the promoter to active state (Kon).

#### Gillespie template

```python
def run_gillespie_telegraph(n0, g0, parameters, t_end, DT):
    """
    Simulates the Random telegraph process using the Gillespie algorithm.

    Parameters:
    n0 : Initial count of species n.
    g0 : Initial promoter state g (0 or 1).
    parameters: Tuple containing transcription rate s, degradation rate k, switching rates k_on and k_off
    t_end : Total simulation time.
    DT: time interval for returning/printing the output

    Returns:
    n_print : Trajectory of species n counts over time.
    g_print : Trajectory of promoter state g over time.
    t_print : Time points for each recorded reaction.
    """
    # Initialize state variables and time
    n = n0   
    g = g0   
    t = 0    
    tp = 0  

    # Initialize output lists
    n_print=[]  
    g_print=[]  
    t_print=[]  

    # Unpack parameters
    s, k, k_on, k_off = parameters 

    while t < t_end: 
        
        # Define the stoichiometric matrix S (also called state-change matrix) and rate vector for each promoter state
        # Notes 4 me: A matrix where rows represent chemical species  and columns represent reactions
        if g == 1: # Promoter is active 
            S = [ [1, -1, 0],
                  [0, 0, -1] 
                ]
            rates = [s, k*n, k_off]
        else: # Promoter is inactive
            S = [ [ -1, 0],
                  [ 0,  1] 
                ]
            rates = [k*n, k_on]

        # Calculate total rate and sample time to next reaction
        rates_tot = sum(rates)
        dt = np.random.exponential(scale=1.0 / rates_tot)
        t += dt

        # Determine which reaction occurs next
        rand = np.random.random()        
        rates_cumsum = np.cumsum(rates)/rates_tot

        i=0
        while rand >= rates_cumsum[i]:
            i=i+1

        # Record output at specified DT intervals (equally spaced output)
        while tp < t and t <= t_end:
            n_print.append(n)
            g_print.append(g)
            t_print.append(tp)
            tp += DT
        
        # Update species n counts and promoter state g
        chosen_react=i
        n += S[0][chosen_react]
        g += S[1][chosen_react]

    return n_print, g_print, t_print
```

```python
def plot_gillespie_telegraph(n, g, t):
    """
    Plots the Random telepgrah process using the Gillespie algorithm.

    Parameters:
    n: Species n counts over time.
    g: Promoter state g over time.
    t: Time points for each recorded reaction.

    """

    fig, axs = plt.subplots(2, figsize=(15, 5), sharex=True)

    # Plotting the number of molecules n over time
    axs[0].plot(t, n, label = 'n(t)', color='darkblue')
    axs[0].set_ylabel('# molecules n')

    # Plotting the promoter activity over time
    axs[1].vlines(x=t, ymin=np.min(g), ymax=np.array(g), colors='teal', lw=2, label='promoter activity')
    axs[1].set_ylim(np.min(g), np.max(g)+0.1)  # Set y-limits to make the promoter states clear (0 and 1)
    axs[1].set_ylabel('Promoter State')
    
    # Common x-label
    plt.xlabel('time [min]')
    
    plt.show()
```

#### Statistical analysis 


Function to compute the statitical informations asked by question 1.2

```python
def statistics_mrna(n_array, g_array):
    """
    Computes statistics for mRNA numbers from simulation arrays.
    
    Parameters:
    - n_array: array of mRNA counts over time
    - g_array: array of promoter states over time (0 or 1)
    
    Returns:
    - pandas DataFrame with formulas applied
    """
    mean_mrna = np.mean(n_array)  
    variance = np.var(n_array)    
    std_dev = np.sqrt(variance)   
    cv = std_dev / mean_mrna      
    p_on = np.mean(g_array)       
    
    # Build DataFrame
    data = {
        "Quantity": [
            "Mean expression level (μ = mean(n))",
            "Variance (σ² = var(n))",
            "Standard deviation (σ = sqrt(Var(n)))",
            "Coefficient of Variation (CV = σ/μ)",
            "Promoter ON probability (p_on = mean(g))",
            "Expected number of molecules (μ)"
        ],
        "Value": [
            mean_mrna,
            variance,
            std_dev,
            cv,
            p_on,
            mean_mrna
        ]
    }
    
    df = pd.DataFrame(data)
    return df
```

####  Helper functions for the Poisson distribution

```python
# Poisson distribution
def Poisson(lam, counts):
    """
    Calculate the Poisson distribution probability for given counts.
    
    Parameters:
    lam: The average number of events (rate parameter).
    counts: Observed counts.

    Returns:
    np.array: Probabilities corresponding to the observed counts.

    Notes:
    - This function uses the log-form of the Poisson probability mass function.
    - The gammaln function is used for calculating the logarithm of the factorial term.

    """

    logp = counts * np.log(lam) - lam - gammaln(counts + 1)
    return np.exp(logp)
```

### Answer 3.1

```python
def run_gillespie_promoter_cycle(n0, g0, parameters, t_end, DT):
    """
    Simulates the Random Telegraph process with a promoter cycle:
    ON (g=1) -> OFF1 (g=2) -> OFF2 (g=3) -> ON (g=1),
    using the Gillespie algorithm.

    Parameters
    ----------
    n0 : int
        Initial number of mRNA molecules.
    g0 : int
        Initial promoter state: 1 = ON, 2 = OFF1, 3 = OFF2.
    parameters : tuple
        (s, k, k_on_off1, k_off1_off2, k_off2_on)
        s            : transcription rate when promoter is ON
        k            : mRNA degradation rate
        k_on_off1    : transition rate ON -> OFF1
        k_off1_off2  : transition rate OFF1 -> OFF2
        k_off2_on    : transition rate OFF2 -> ON
    t_end : float
        Total simulation time.
    DT : float
        Time step for recording the output.

    Returns
    -------
    n_print : list
        Trajectory of mRNA copy numbers.
    g_print : list
        Trajectory of promoter states (1, 2, or 3).
    t_print : list
        Time points corresponding to the recorded trajectories.
    """

    # Initialize state variables and time 
    n = n0          # current mRNA count
    g = g0          # current promoter state
    t = 0.0         # current time
    tp = 0.0        # time of next recording

    # Initialize output trajectories 
    n_print = []
    g_print = []
    t_print = []

    # Unpack parameters 
    s, k, k_on_off1, k_off1_off2, k_off2_on = parameters

    # Gillespie loop 
    while t < t_end:

        # Define stoichiometric matrix S and reaction rates
        # depending on the current promoter state g.
        #
        # We keep two "species":
        #  - row 0: mRNA counts n
        #  - row 1: promoter state g (encoded as 1, 2, or 3)
        #
        # Each column in S corresponds to one reaction and contains
        # the change of each species when that reaction fires.

        if g == 1:  # ON state
            # Reactions:
            #  1. Transcription:      n -> n + 1        (rate s)
            #  2. Degradation:       n -> n - 1        (rate k * n)
            #  3. ON -> OFF1:        g: 1 -> 2  (Δg = +1) (rate k_on_off1)
            S = np.array([[ 1, -1,  0],
                          [ 0,  0,  1]])
            rates = np.array([s, k * n, k_on_off1])

        elif g == 2:  # OFF1 state
            # Reactions:
            #  1. Degradation:       n -> n - 1          (rate k * n)
            #  2. OFF1 -> OFF2:      g: 2 -> 3 (Δg = +1) (rate k_off1_off2)
            S = np.array([[-1,  0],
                          [ 0,  1]])
            rates = np.array([k * n, k_off1_off2])

        elif g == 3:  # OFF2 state
            # Reactions:
            #  1. Degradation:       n -> n - 1           (rate k * n)
            #  2. OFF2 -> ON:        g: 3 -> 1 (Δg = -2)  (rate k_off2_on)
            S = np.array([[-1,  0],
                          [ 0, -2]])
            rates = np.array([k * n, k_off2_on])

        else:
            # Just in case, to avoid invalid states
            raise ValueError(f"Unknown promoter state g = {g}")

        # If there are no reactions possible (all rates = 0), stop the simulation
        rates_tot = np.sum(rates)
        if rates_tot <= 0:
            break

        # Sample time to the next reaction 
        dt = np.random.exponential(scale=1.0 / rates_tot)
        t_next = t + dt

        # Record output at fixed DT intervals 
        while tp < t_next and tp <= t_end:
            n_print.append(n)
            g_print.append(g)
            t_print.append(tp)
            tp += DT

        # If we are already beyond the end time, stop
        if t_next > t_end:
            t = t_end
            break

        # Choose which reaction occurs 
        rand = np.random.rand()
        rates_cumsum = np.cumsum(rates) / rates_tot
        chosen_react = np.searchsorted(rates_cumsum, rand)

        # Update state according to chosen reaction 
        n += S[0, chosen_react]
        g += S[1, chosen_react]

        # Prevent negative mRNA counts
        if n < 0:
            n = 0

        t = t_next

    return n_print, g_print, t_print
```

```python
def plot_gillespie_promoter_cycle(n, g, t):
    """
    Plots the Random Telegraph process with a promoter cycle.

    Parameters
    ----------
    n : array-like
        mRNA copy number over time.
    g : array-like
        Promoter state over time (1 = ON, 2 = OFF1, 3 = OFF2).
    t : array-like
        Time points.
    """

    fig, axs = plt.subplots(2, figsize=(15, 5), sharex=True)

    # Plot mRNA copy number 
    axs[0].plot(t, n, label='n(t)')
    axs[0].set_ylabel('# molecules n')
    axs[0].set_title('Promoter cycle with two OFF states')
    axs[0].legend()

    # Plot promoter state as vertical lines 
    axs[1].vlines(x=t,
                  ymin=np.min(g),
                  ymax=np.array(g),
                  lw=2,
                  label='promoter state (1=ON, 2=OFF1, 3=OFF2)')
    axs[1].set_ylim(np.min(g), np.max(g) + 0.1)
    axs[1].set_ylabel('Promoter state')
    axs[1].legend()

    axs[1].set_xlabel('time [min]')
    plt.tight_layout()
    plt.show()
```

```python
# Initial conditions 
n_0 = 0      # start with no mRNA (system will relax to steady state)
g_0 = 1      # start with promoter ON
DT = 0.1     # recording time step [min]
t_end = 10000  # total simulation time [min]

# Parameters for the promoter cycle 
s = 20                        # transcription rate when ON
k = 0.02                      # degradation rate
k_on_off1   = 1/15            # ON -> OFF1
k_off1_off2 = 1/60            # OFF1 -> OFF2
k_off2_on   = 1/60            # OFF2 -> ON

parameters_cycle = (s, k, k_on_off1, k_off1_off2, k_off2_on)

# Run simulation 
n_cycle, g_cycle, t_cycle = run_gillespie_promoter_cycle(
    n_0, g_0, parameters_cycle, t_end, DT
)

# Plot trajectories 
plot_gillespie_promoter_cycle(n_cycle, g_cycle, t_cycle)

```

### Interpretation of the promoter-cycle simulations

With the promoter cycle **ON → OFF₁ → OFF₂ → ON**, the system must pass through **two sequential OFF states** before returning to the ON state. This introduces a **refractory period**, meaning that once the promoter turns OFF, it remains inactive for a well-defined amount of time before becoming transcriptionally active again.

The effect on the dynamics is clear:

* **During ON states (g = 1)**, transcription occurs at a high rate (`s = 20`), leading to **rapid production of mRNA** and the appearance of sharp **transcriptional bursts**.
* **During OFF₁ and OFF₂ states (g = 2, 3)**, transcription stops entirely while degradation continues. As a result, the mRNA count decreases back toward zero.
* The presence of two OFF states makes the OFF periods **longer and more regular** than in the simple two-state telegraph model.

Overall, the promoter cycle generates **more separated, more periodic bursts** with well-defined silent intervals.
Even if the mean expression remains comparable to simpler models (depending on the choice of rates), the refractory period leads to:

* **Higher cell-to-cell and temporal variability** in mRNA numbers (larger variance and coefficient of variation),
* **More pronounced burstiness**,
* **Longer quiescent phases** between bursts.

This behaviour captures biological observations in mammalian promoters, where transcription often requires multiple steps before reactivation, naturally producing bursty expression with refractory periods.

### OFF-interval duration analysis

The durations of **OFF₁** and **OFF₂** follow **exponential distributions**, each governed by a single memoryless Markov transition (rate = 1/60). These show many short intervals and a long tail.  
The **total OFF duration** (OFF₁ + OFF₂) is the **sum of two exponentials**, producing an **Erlang/Gamma distribution (k = 2)**. This distribution is not maximal at zero but displays a clear peak around 60–80 minutes, reflecting a more regular and less noisy OFF period.  
Biologically, this corresponds to a **refractory mechanism**: the promoter must complete two sequential OFF steps before reactivation, leading to more predictable OFF intervals and more periodic transcriptional bursts.

### Mathematical and biological explanation of the OFF-interval distribution

Mathematically, each OFF step is an **exponential waiting time**, because promoter transitions follow a memoryless Markov process. The total OFF interval is therefore the **sum of several exponential variables**. The sum of *k* identical exponentials is a **Gamma/Erlang distribution with shape k**, which has:
- probability density = 0 at time 0,
- a peak at $(k-1)\theta\$,
- decreasing relative noise as k increases.

Biologically, this reflects the fact that promoter reactivation requires several **sequential biochemical steps** (chromatin opening, cofactor binding, polymerase loading…). Each step has its own stochastic duration. Adding more OFFₙ states makes the OFF period **more deterministic** and less noisy: the distribution becomes more sharply peaked and approaches a **Gaussian** by the central limit theorem. This yields **more regular, periodic transcriptional bursts** with a well-defined refractory period.
